### **Завдання 1**

```python
from google.colab import files
import io
import pandas as pd

uploaded = files.upload()
p = pd.read_csv(io.BytesIO(uploaded["points.csv"]))
```

### **Завдання 2**

```python
p.info()
p.describe()
```

### **Завдання 3**

```python
variances = p.var()
correlation_matrix = p.corr()

print("Дисперсії для кожної координати:")
print(variances)

print("\nКореляційна матриця:")
print(correlation_matrix)
```

### **Завдання 4**

```python
import plotly.express as px
```

### **Завдання 5-7**

```python
fig = px.scatter_3d(data_frame=p, x='x', y='y', z='z')
fig.update_traces(marker={'size': 3})
fig.update_layout(scene={'aspectmode': 'data'})
```

### **Завдання 9**

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(p)
```

### **Завдання 10**

```python
p_trans_array = pca.transform(p)

p_trans = pd.DataFrame(data=p_trans_array, columns=['c1', 'c2', 'c3'])

p_trans.head()
```

```python
p.head()
```

### **Завдання 11**

```python
print("Статистика нових координат:")
print(p_trans.describe())

variances_trans = p_trans.var()

correlation_matrix_trans = p_trans.corr()

print("\nДисперсії для нових координат:")
print(variances_trans)

print("\nКореляційна матриця для нових координат:")
print(correlation_matrix_trans)
```

### **Завдання 12**

```python
fig_trans = px.scatter_3d(data_frame=p_trans, x='c1', y='c2', z='c3')
fig_trans.update_traces(marker={'size': 3})
fig_trans.update_layout(scene={'aspectmode': 'data'})
```

### **Завдання 13**

```python
p_trans_dimred = p_trans.copy()
p_trans_dimred['c3'] = 0
```

### **Завдання 14**

```python
fig_trans.add_scatter3d(x=p_trans_dimred['c1'],
                        y=p_trans_dimred['c2'],
                        z=p_trans_dimred['c3'],
                        mode='markers',
                        marker_size=3)
```

### **Завдання 15**

```python
p_dimred_array = pca.inverse_transform(p_trans_dimred)
p_dimred = pd.DataFrame(data=p_dimred_array, columns=['x', 'y', 'z'])
```

### **Завдання 16**

```python
fig.add_scatter3d(x=p_dimred['x'],
                  y=p_dimred['y'],
                  z=p_dimred['z'],
                  mode='markers',
                  marker_size=3)
```
